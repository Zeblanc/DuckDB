```python
import duckdb as db
import pandas as pd
import numpy as np

conn = db.connect()
```

```python
conn.sql("CREATE TABLE IF NOT EXISTS counties AS SELECT * FROM read_csv_auto('https://raw.githubusercontent.com/roualdes/data/refs/heads/master/county.csv')")
```

```python
conn.sql("SELECT * FROM counties")
```

```python
mean_age = conn.sql('''
                        SELECT AVG(age_under_18) FROM COUNTIES
                        ''').fetchone()[0]

print(f"Duckdb Mean age under 18: {mean_age}")
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/roualdes/data/refs/heads/master/county.csv')

mean_age_df = df["age_under_18"].mean()

print(f"Pandas Mean age under 18: {mean_age_df}")
```

```python
mean_age_np = np.mean(df["age_under_18"])

print(f"Numpy Mean age under 18: {mean_age_np}")
```

```python
mean_age_db = conn.sql('''
                           SELECT age_under_18 FROM COUNTIES
                           ''').fetchnumpy()

mean_age_db

print(f"DuckDB with Numpy {np.mean(mean_age_db['age_under_18'])}")
```

```python
conn.close()
```
